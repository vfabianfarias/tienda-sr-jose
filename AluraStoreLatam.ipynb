### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular ingresos totales por tienda sumando la columna 'Precio'
ingresos_tienda1 = tienda["Precio"].sum()
ingresos_tienda2 = tienda2["Precio"].sum()
ingresos_tienda3 = tienda3["Precio"].sum()
ingresos_tienda4 = tienda4["Precio"].sum()

# Mostrar resultados como enteros sin decimales
print("Ingresos totales por tienda (suma de columna 'Precio'):")
print(f"Tienda 1: ${int(ingresos_tienda1):,}")
print(f"Tienda 2: ${int(ingresos_tienda2):,}")
print(f"Tienda 3: ${int(ingresos_tienda3):,}")
print(f"Tienda 4: ${int(ingresos_tienda4):,}")
```

```python
import matplotlib.pyplot as plt

# Datos de ingresos como enteros
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [
    int(ingresos_tienda1),
    int(ingresos_tienda2),
    int(ingresos_tienda3),
    int(ingresos_tienda4)
]

# Crear gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos)
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos ($)")
plt.xlabel("Tienda")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Agrupar por categoría y contar la cantidad de ventas (1 venta por fila)
ventas_cat1 = tienda["Categoría del Producto"].value_counts()
ventas_cat2 = tienda2["Categoría del Producto"].value_counts()
ventas_cat3 = tienda3["Categoría del Producto"].value_counts()
ventas_cat4 = tienda4["Categoría del Producto"].value_counts()

# Mostrar resultados
print("Tienda 1 - Ventas por categoría:")
print(ventas_cat1, '\n')

print("Tienda 2 - Ventas por categoría:")
print(ventas_cat2, '\n')

print("Tienda 3 - Ventas por categoría:")
print(ventas_cat3, '\n')

print("Tienda 4 - Ventas por categoría:")
print(ventas_cat4)
```

```python
# Lista de datos y nombres
ventas = [ventas_cat1, ventas_cat2, ventas_cat3, ventas_cat4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Crear subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
axes = axes.flatten()

for i in range(4):
    ventas[i].plot(kind="bar", ax=axes[i])
    axes[i].set_title(f"Ventas por categoría - {nombres_tiendas[i]}")
    axes[i].set_ylabel("Cantidad de ventas")
    axes[i].set_xlabel("Categoría")
    axes[i].tick_params(axis='x', labelrotation=35)  # <--- rotación más clara
    axes[i].set_xticklabels(axes[i].get_xticklabels(), ha='right')  # <--- alineación derecha

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular calificación promedio por tienda
calif_prom_t1 = tienda["Calificación"].mean()
calif_prom_t2 = tienda2["Calificación"].mean()
calif_prom_t3 = tienda3["Calificación"].mean()
calif_prom_t4 = tienda4["Calificación"].mean()

# Mostrar resultados redondeados
print("Calificación promedio por tienda:")
print(f"Tienda 1: {round(calif_prom_t1, 2)} estrellas")
print(f"Tienda 2: {round(calif_prom_t2, 2)} estrellas")
print(f"Tienda 3: {round(calif_prom_t3, 2)} estrellas")
print(f"Tienda 4: {round(calif_prom_t4, 2)} estrellas")
```

```python
# Nombres y valores redondeados
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
calificaciones = [
    round(calif_prom_t1, 2),
    round(calif_prom_t2, 2),
    round(calif_prom_t3, 2),
    round(calif_prom_t4, 2)
]

# Crear gráfico
plt.figure(figsize=(8, 5))
bars = plt.bar(tiendas, calificaciones, color="skyblue")
plt.ylim(0, 5)  # Límite de escala para calificaciones
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación (de 1 a 5)")
plt.xlabel("Tienda")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Agregar etiquetas sobre las barras
for bar, value in zip(bars, calificaciones):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, str(value),
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Contar cuántas veces se vendió cada producto por tienda
productos_t1 = tienda["Producto"].value_counts()
productos_t2 = tienda2["Producto"].value_counts()
productos_t3 = tienda3["Producto"].value_counts()
productos_t4 = tienda4["Producto"].value_counts()

# Tienda 1
producto_mas_vendido_t1 = productos_t1.idxmax()
ventas_max_t1 = productos_t1.max()
producto_menos_vendido_t1 = productos_t1.idxmin()
ventas_min_t1 = productos_t1.min()

# Tienda 2
producto_mas_vendido_t2 = productos_t2.idxmax()
ventas_max_t2 = productos_t2.max()
producto_menos_vendido_t2 = productos_t2.idxmin()
ventas_min_t2 = productos_t2.min()

# Tienda 3
producto_mas_vendido_t3 = productos_t3.idxmax()
ventas_max_t3 = productos_t3.max()
producto_menos_vendido_t3 = productos_t3.idxmin()
ventas_min_t3 = productos_t3.min()

# Tienda 4
producto_mas_vendido_t4 = productos_t4.idxmax()
ventas_max_t4 = productos_t4.max()
producto_menos_vendido_t4 = productos_t4.idxmin()
ventas_min_t4 = productos_t4.min()

# Mostrar en texto
print("Productos más vendidos:")
print(f"Tienda 1: {producto_mas_vendido_t1} ({ventas_max_t1} ventas)")
print(f"Tienda 2: {producto_mas_vendido_t2} ({ventas_max_t2} ventas)")
print(f"Tienda 3: {producto_mas_vendido_t3} ({ventas_max_t3} ventas)")
print(f"Tienda 4: {producto_mas_vendido_t4} ({ventas_max_t4} ventas)\n")

print("Productos menos vendidos:")
print(f"Tienda 1: {producto_menos_vendido_t1} ({ventas_min_t1} venta)")
print(f"Tienda 2: {producto_menos_vendido_t2} ({ventas_min_t2} venta)")
print(f"Tienda 3: {producto_menos_vendido_t3} ({ventas_min_t3} venta)")
print(f"Tienda 4: {producto_menos_vendido_t4} ({ventas_min_t4} venta)")
```

```python
import numpy as np

# Datos base
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

ventas_max = [ventas_max_t1, ventas_max_t2, ventas_max_t3, ventas_max_t4]
ventas_min = [ventas_min_t1, ventas_min_t2, ventas_min_t3, ventas_min_t4]

nombres_max = [producto_mas_vendido_t1, producto_mas_vendido_t2, producto_mas_vendido_t3, producto_mas_vendido_t4]
nombres_min = [producto_menos_vendido_t1, producto_menos_vendido_t2, producto_menos_vendido_t3, producto_menos_vendido_t4]

x = np.arange(len(tiendas))
ancho = 0.35

plt.figure(figsize=(10, 6))

# Barras
barras_max = plt.bar(x - ancho/2, ventas_max, width=ancho, label="Más vendido", color="green")
barras_min = plt.bar(x + ancho/2, ventas_min, width=ancho, label="Menos vendido", color="red")

# Ejes y título
plt.title("Productos más y menos vendidos por tienda")
plt.xlabel("Tienda")
plt.ylabel("Cantidad de ventas")
plt.xticks(x, tiendas)

# Mover leyenda abajo
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=2, frameon=False)

# Etiquetas sobre las barras
for i in range(len(tiendas)):
    # Más vendido
    plt.text(barras_max[i].get_x() + barras_max[i].get_width()/2,
             barras_max[i].get_height() + 0.5,
             f"{nombres_max[i]}\n({ventas_max[i]})",
             ha='center', va='bottom', fontsize=9, rotation=45)

    # Menos vendido
    plt.text(barras_min[i].get_x() + barras_min[i].get_width()/2,
             barras_min[i].get_height() + 0.5,
             f"{nombres_min[i]}\n({ventas_min[i]})",
             ha='center', va='bottom', fontsize=9, rotation=45)

plt.tight_layout()
plt.show()
```

# 7. Informe final

## Introducción

El presente informe tiene como objetivo evaluar el rendimiento de las cuatro tiendas que componen la cadena **Alura Store**, con el fin de determinar cuál de ellas debería vender el Sr. Juan para enfocarse en un nuevo emprendimiento. La decisión se basa en el análisis de diversos factores clave: **ingresos totales, productos y categorías más y menos vendidos, calificaciones promedio de clientes, costos de envío y métodos de pago utilizados**.

Los datos fueron analizados utilizando **Python con Pandas y Matplotlib**, y visualizados mediante gráficos adecuados para facilitar la interpretación. A continuación, se presentan los hallazgos principales.

---

## Desarrollo del análisis

### 1. Ingresos totales por tienda

El análisis de ingresos totales muestra claras diferencias entre las tiendas. **Tienda 3** se posiciona como la más rentable, mientras que **Tienda 4** obtiene los ingresos más bajos. Esta métrica es fundamental, ya que refleja la eficiencia en ventas.

### 2. Distribución de ventas por categoría

Las categorías más populares varían entre tiendas. Sin embargo, **Tienda 1** y **Tienda 3** destacan por tener una mayor concentración de ventas en categorías con alta demanda, como **Electrodomésticos** y **Electrónicos**. **Tienda 4** muestra una distribución más débil y dispersa.

### 3. Calificación promedio de los clientes

Las calificaciones de los clientes reflejan su experiencia de compra. **Tienda 2** presenta la mejor calificación promedio, seguida por **Tienda 1**. **Tienda 3**, a pesar de sus ingresos altos, tiene una calificación más baja, lo que podría indicar una oportunidad de mejora en atención o logística. **Tienda 4** nuevamente se encuentra en desventaja con calificaciones modestas.

### 4. Productos más y menos vendidos

Al observar los productos individuales, **Tienda 1** y **Tienda 3** presentan una clara presencia de productos líderes en ventas. **Tienda 4** tiene un bajo volumen tanto en sus productos más vendidos como en los menos vendidos, lo que evidencia una falta de protagonismo en el catálogo general.

### 5. Costo de envío promedio

Los costos de envío son un componente importante de la percepción del cliente. **Tienda 3** mantiene un costo promedio competitivo. **Tienda 4**, en cambio, posee uno de los **costos de envío más altos**, lo que impacta directamente la satisfacción del cliente y la conversión de ventas.

### 6. Métodos de pago

El gráfico de métodos de pago muestra una buena diversidad en todas las tiendas, aunque **Tienda 4** presenta menor variedad y volumen, lo que también puede estar relacionado con su bajo rendimiento general.

---

## Conclusión y recomendación

Con base en los datos analizados, la tienda que **presenta consistentemente el rendimiento más débil es la Tienda 4**. Esta tienda tiene:

- Los **ingresos más bajos** de toda la cadena.
- Una **distribución de ventas por categoría poco destacada**.
- **Calificaciones promedio inferiores** en comparación con las otras tiendas.
- **Costos de envío más altos**, lo que puede desalentar a los compradores.
- Productos con **baja rotación** y sin un líder claro en ventas.
- Menor diversidad y volumen en los **métodos de pago** registrados.

Por estas razones, se recomienda que el **Sr. Juan considere vender la Tienda 4**, ya que presenta debilidades en casi todos los indicadores analizados. Esta decisión le permitirá concentrar sus recursos en las tiendas con mayor potencial de crecimiento y rentabilidad.


# 8. Actividad Extra – Análisis Geográfico de Ventas

Como complemento al análisis de rendimiento de las tiendas, se exploró la distribución geográfica de las ventas utilizando las coordenadas de latitud y longitud presentes en los datos. El objetivo fue identificar posibles patrones espaciales que influyan en el desempeño de las tiendas, visualizando la ubicación de las compras y su concentración por región.

```python
import pandas as pd
import folium

# Cargar y etiquetar datos (igual que antes)
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda1["Tienda"] = "Tienda 1"
tienda2 = pd.read_csv(url2)
tienda2["Tienda"] = "Tienda 2"
tienda3 = pd.read_csv(url3)
tienda3["Tienda"] = "Tienda 3"
tienda4 = pd.read_csv(url4)
tienda4["Tienda"] = "Tienda 4"

datos = pd.concat([tienda1, tienda2, tienda3, tienda4])

# Agrupar por coordenadas y contar cantidad de ventas
ventas_por_ubicacion = datos.groupby(["lat", "lon"]).size().reset_index(name="cantidad")

# Crear mapa base
mapa_resumido = folium.Map(location=[4.6, -74.1], zoom_start=6)

# Agregar marcador único por ubicación, con el total de ventas
for _, row in ventas_por_ubicacion.iterrows():
    folium.CircleMarker(
        location=[row["lat"], row["lon"]],
        radius=5,
        color="blue",
        fill=True,
        fill_color="blue",
        fill_opacity=0.6,
        popup=f'Ventas en esta ubicación: {int(row["cantidad"])}'
    ).add_to(mapa_resumido)

# Mostrar mapa
mapa_resumido
```
